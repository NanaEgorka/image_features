```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
folder_path = 'vae_microstructure/data/data_numpy_format'
image_pathes = os.listdir(folder_path)

carbn_pathes = list(filter(lambda x : x.startswith('carbn'), image_pathes))
pearl_pathes = list(filter(lambda x : x.startswith('pearl'), image_pathes))
spher_pathes = list(filter(lambda x : x.startswith('spher'), image_pathes))
assert len(image_pathes) == len(carbn_pathes) + len(pearl_pathes) + len(spher_pathes)
```

```python
def display_image_class(folder_path, class_pathes):
    shapes = []
    fig, ax = plt.subplots(2, 3, figsize=(20, 10))
    for i, image_path in enumerate(class_pathes):
        image = np.load(os.path.join(folder_path, image_path))
        shapes.append(image.shape) 
        if i < 6:
            ax[i // 3][i % 3].imshow(image, cmap='gray')
    print('Unique shapes:', set(shapes))
```

```python
display_image_class(folder_path, carbn_pathes)
```

```python
display_image_class(folder_path, pearl_pathes)
```

```python
display_image_class(folder_path, spher_pathes)
```
